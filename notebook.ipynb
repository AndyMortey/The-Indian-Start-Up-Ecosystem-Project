```python
print("USERNAME:", username)
print("PASSWORD:", password)

# Construct the connection string
connection_string = f"DRIVER=ODBC Driver 17 for SQL Server;SERVER={server};DATABASE={database};UID={username};PWD={password};"
# Connect to the database
try:
    connection = pyodbc.connect(connection_string)
    print("Connection successful!")
except Exception as e:
    print("Error:", e)
```

```python
# Specify the SQL queries to extract data from the tables
query_2020 = "SELECT * FROM dbo.LP1_startup_funding2020"
query_2021 = "SELECT * FROM dbo.LP1_startup_funding2021"
```

```python
import warnings
import pandas as pd

# Suppress warnings
warnings.filterwarnings('ignore')

# Create a cursor from the connection
with connection.cursor() as cursor:
    # Execute the queries and fetch data into Pandas DataFrames
    df_2020 = pd.read_sql_query(query_2020, connection)
    df_2021 = pd.read_sql_query(query_2021, connection)
```

Null Hypothesis (H0): The location of a startup's headquarters has no significant impact on the amount of funding it receives.

Alternative Hypothesis (H1): The location of a startup's headquarters is associated with a significant difference in the amount of funding it receives.

Analytical Questions:
1. Geographical Impact:

Is there a correlation between the location of a startup (HeadQuarter) and the funding it receives?
Are startups in specific regions more likely to receive higher funding?

2. Sector Analysis:

Which sectors (AgriTech, EdTech, Hygiene management, etc.) attract more funding on average?
Are certain sectors more likely to have successful startups?

3. Company Age and Funding:

Is there a relationship between the founding year of a company and the funding it receives?
Do newer startups tend to attract more funding?

4. Sector-wise What It Does Analysis:

How do the "What_it_does" descriptions vary across different sectors?
Are there common themes or patterns in what startups in specific sectors aim to achieve?

5. Success Factors:
